### Setup

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = pd.read_csv('https://raw.githubusercontent.com/Thinkful-Ed/curric-data-001-data-sets/master/iris/iris.data.csv',
                  names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'], header = None)
```

```python
iris.head()
```

```python
sns.pairplot(iris, diag_kind="kde", hue="class", markers=["o", "s", "D"])
```

```python
def knneucl(sep_length, sep_width, k):
    def f(x):
        return np.sqrt((sep_length - x['sepal_length'])**2 + (sep_width - x['sepal_width'])**2)
    dist = iris.apply(f, axis = 1)
    return Counter(iris.iloc[dist.nsmallest(k).index]['class'])
```

```python
knneucl(5.0, 4.0, 10)
```

```python
knneucl(5.2, 3.0, 10)
```
